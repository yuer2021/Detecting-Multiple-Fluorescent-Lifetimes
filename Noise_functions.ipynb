```python
def add_poission_noise(decay):
  # Signal dependent noise (shot noise)
  plt.plot(decay)
  decay = decay + np.random.poission(decay)
  plt.plot(decay)
  plt.show()
  return (decay)

def add_gaussian_noise(decay, noise_mean, noise_spread):
  # Read noise from electronics
  # In general use same parameteres in entire image
  # Start with small values
  plt.plot(decay)
  decay = decay + np.random.normal(noise_mean, noise_spread)
  plt.plot(decay)
  plt.show()
  return (decay)
```
